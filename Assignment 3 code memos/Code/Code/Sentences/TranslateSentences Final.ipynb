```python
import pandas as pd


df_sentences = pd.read_csv("GeneratedSentences.csv")
df = pd.read_csv('LexiconPOSClean.csv',sep=";")
from googletrans import Translator, constants
from pprint import pprint
translator = Translator()
# Create translation and scoring dictionaries from the dataset

translation_lexique = dict(zip(df['french'].str.lower(), df['ciluba'])) # French to ciluba
lexique = dict(zip(df['ciluba'].str.lower(), df['score']))
french_lexicon = dict(zip(df['English'],df['french'].str.lower())) # english to Afrikaans
afrikaans_lexicon = dict(zip(df['English'],df['Afrikaans'].str.lower())) # english to Afrikaans
zulu_lexicon = dict(zip(df['English'],df['Zulu'].str.lower())) # english to Afrikaans
xhosa_lexicon = dict(zip(df['English'],df['Xhosa'].str.lower())) # english to Afrikaans
#https://www.twinword.com/blog/interpreting-the-score-and-ratio-of-sentiment/

try:
    df_translatedSentences
except NameError:
    df_translatedSentences = pd.DataFrame(columns=['French','English','Afrikaans','Zulu','Xhosa'])

# Do it recursively
def translate_sentence(text,src,lexicon):
    words = text.lower().split()
    # remove full stops
    #print(words)
    #https://stackoverflow.com/questions/21877729/remove-fullstop-commas-quotation-from-list-in-python
    words = [word.replace(".", "") for word in words]
    # check word then translate and add to lexicon for english word.
    print("words no full stop",words)
    list = []
    translated_text = ''
    for word in words:
        if word in lexicon:
            #print("in",word)
            add_word = lexicon.get(word, word)
            print("in Lexicon:",add_word)
            translated_text += ' ' + add_word
        else:
            #print("out",word)
            add_word = Translate(word,src)
            #print(add_word)
            translated_text += ' ' + add_word
    #https://stackoverflow.com/questions/42451442/how-to-convert-a-list-of-words-into-sentences-in-python
    #print(list)
    #translated_text = [' '.join(i) for i in list]
    print(translated_text)
    #https://stackoverflow.com/questions/959215/how-do-i-remove-leading-whitespace-in-python
    return translated_text.lstrip()

df_translatedSentences['French'] = (df_sentences['English']).apply(translate_sentence,src="fr",lexicon=french_lexicon)
df_translatedSentences['Afrikaans'] = (df_sentences['English']).apply(translate_sentence,src="af",lexicon=afrikaans_lexicon)
df_translatedSentences['Zulu'] = (df_sentences['English']).apply(translate_sentence,src="zu",lexicon=zulu_lexicon)
df_translatedSentences['Xhosa'] = (df_sentences['English']).apply(translate_sentence,src="xh",lexicon=xhosa_lexicon)
df_sentences.to_csv('TranslatedSentences.csv',encoding='utf-8-sig')

def Translate(word,dest):
    print("In Translate",word)
    translation = translator.translate(word,dest=dest)
    print("In Translate",translation.text)
    return translation.text
```
